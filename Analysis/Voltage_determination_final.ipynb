```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress, moyal
from scipy.optimize import curve_fit

files = [
    '20241111_PM18astrigger_noabs_0deg_1800V.txt',
    '20241111_PM18astrig_noabs_0deg_1850V.txt',
    '20241111_PM18astrigger_noabs_0deg_1900V.txt',
    '20241111_PM18astrigger_noabs_0deg_1950V.txt',
    '20241111_PM18astrigger_noabs_0deg_2000V.txt'
]

voltages = [-1800, -1850, -1900, -1950, -2000]

channels = np.array([
    np.array([283, 308, 358, 442]),
    np.array([141, 159, 201, 282]),
    np.array([144, 162, 204, 287]),
    np.array([147, 165, 208, 291]),
    np.array([139, 157, 199, 281]),
    np.array([133, 151, 194, 276]),
    np.array([143, 161, 203, 285]),
    np.array([132, 151, 192, 275])
])

Delay = np.array([2, 4, 8, 16])

channel_fit_params = [linregress(channel, Delay)[:2] for channel in channels]

fig, axs = plt.subplots(8, 5, figsize=(25, 30))

for j, file in enumerate(files):
    ADC = np.genfromtxt(file, dtype=None, delimiter='\t', usecols=(1,2,3,4,5,6,7,8))

    for i in range(ADC.shape[1]):
        ax_adc_raw = axs[i, j]
        counts, bin_edges = np.histogram(ADC[:, i], range=(100, 1100), bins=200, density=True)
        bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

        try:
            popt, _ = curve_fit(moyal.pdf, bin_centers, counts, p0=[500, 50])
            loc, scale = popt

            ax_adc_raw.hist(ADC[:, i], range=(100, 1100), bins=200, color='orange', alpha=0.6, density=True)
            x_fit = np.linspace(100, 1100, 500)
            y_fit = moyal.pdf(x_fit, *popt)
            ax_adc_raw.plot(x_fit, y_fit, 'r-', label=f'Fit: loc={loc:.2f}, scale={scale:.2f}')
        except RuntimeError:
            ax_adc_raw.hist(ADC[:, i], range=(100, 1100), bins=200, color='orange', alpha=0.6, density=True)
            ax_adc_raw.text(0.5, 0.8, 'Fit Failed', transform=ax_adc_raw.transAxes, fontsize=10, color='red', ha='center')

        ax_adc_raw.set_title(f'PMT {i+1} @ {voltages[j]}V')
        ax_adc_raw.legend()

plt.tight_layout()
plt.savefig('ADC_Plots.png')
plt.show()
```

```python
# For the report
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moyal
from scipy.optimize import curve_fit

files = [
    '20241111_PM18astrigger_noabs_0deg_1800V.txt',
    '20241111_PM18astrig_noabs_0deg_1850V.txt',
    '20241111_PM18astrigger_noabs_0deg_1900V.txt',
    '20241111_PM18astrigger_noabs_0deg_1950V.txt',
    '20241111_PM18astrigger_noabs_0deg_2000V.txt'
]

voltages = [-1800, -1850, -1900, -1950, -2000]

for j, file in enumerate(files):
    ADC = np.genfromtxt(file, dtype=None, delimiter='\t', usecols=(1,2,3,4,5,6,7,8))

    fig_adc, axs_adc = plt.subplots(2, 4, figsize=(16, 8))
    axs_adc = axs_adc.flatten()
    fig_adc.suptitle(f'ADC Distributions at {voltages[j]}V', fontsize=16)

    for i in range(ADC.shape[1]):
        counts, bin_edges = np.histogram(ADC[:, i], range=(100, 1100), bins=100, density=True)
        bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

        try:
            popt, _ = curve_fit(moyal.pdf, bin_centers, counts, p0=[500, 50])
            loc, scale = popt
            x_fit = np.linspace(100, 1100, 500)
            y_fit = moyal.pdf(x_fit, *popt)
            axs_adc[i].plot(x_fit, y_fit, 'r-', label=f'Fit: loc={loc:.2f}, scale={scale:.2f}')
        except RuntimeError:
            axs_adc[i].text(0.5, 0.8, 'Fit Failed', transform=axs_adc[i].transAxes, fontsize=12, color='red', ha='center')

        axs_adc[i].hist(ADC[:, i], range=(100, 1100), bins=100, color='orange', alpha=0.6, density=True, label=f'ADC PMT {i+1}')
        axs_adc[i].set_title(f'ADC PMT {i+1}', fontsize=14)
        axs_adc[i].legend(fontsize=12)
        axs_adc[i].tick_params(axis='both', which='major', labelsize=12)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.savefig(f'ADC_Distributions_{voltages[j]}V.png', dpi=300)
    plt.show()
```
